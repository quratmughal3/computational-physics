# LINEARIZATION OF NON-LINEAR RELATIONSHIP

# linearization of a power equation

```python
import numpy as np
from numpy import log as ln
x = []
y = [] 
X = []
Y = []
X_Y = []
X_square = []
n = int(input("enter total number of elements : "))
for i in range(0,n):
    x.append(int(input("enter elements of x: ")))
    y.append(int(input("enter elements of y: ")))
    a  = ln(x)
    b =  ln(y)
X = np.sum(a)
Y = np.sum(b)
X_Y.append(np.sum(a*b)) 
def power(lst,power):
    return [i**power for i in lst]
X_square.append(np.sum(power(a,2))) 
B = (np.multiply(n,X_Y) - X*Y) / (np.multiply(n,X_square) - (X**2))
A = (Y - B*X)/n
a = np.e**A
print(B , 
      a 
     )

import matplotlib.pyplot as plt
y_new = a*x **B
print("==================graph of untransformed data==============")
plt.plot(ln(x), y , "o" , linestyle = "--")
plt.grid(True , linestyle = ":")
plt.show()
print("===================graph of transformed data===============")
plt.plot(ln(x),y_new , "o" , linestyle = "dotted")
plt.grid(True , linestyle = ":")
plt.show()
```
